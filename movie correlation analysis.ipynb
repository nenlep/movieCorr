```python
#libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure
from IPython.display import display

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)
```

```python
#load dataset into dataframe
df1 = pd.read_csv(r'C:\Users\Owner\Downloads\movies.csv')
df1.head()
```

```python
#check for nan values
for col in df1.columns:
    miss_pct = np.mean(df1[col].isnull())
    print('{}={}%'.format(col,miss_pct))
```

```python
#drop rows with na values
df1 = df1.dropna()
```

```python
for col in df1.columns:
    miss_pct = np.mean(df1[col].isnull())
    print('{}={}%'.format(col,miss_pct))
```

```python
df1.dtypes
```

```python
#change data types
df1['gross'] = df1['gross'].astype(np.int64)
df1['runtime'] = df1['runtime'].astype(np.int64)
df1['votes'] = df1['votes'].astype(np.int64)
df1['budget'] = df1['budget'].astype(np.int64)
display(df1)
```

```python
#sort dataframe by gross sales
df1.sort_values(by=['gross'],inplace=False, ascending=False)
```

```python
uniq = df1['company'].unique()
uniq = uniq.tolist()
uniq
```

```python
#scatterplot budget v gross rev.
plt.scatter(x=df1['budget'],y=df1['gross'])
plt.title('scatter plot for movie budget v gross revenue')
plt.xlabel('movie budget')
plt.ylabel('gross revenue')
plt.show
```

```python
#budget v gross regression plot
sns.regplot(x='budget',y='gross',data=df1,scatter_kws={'color':'red'},line_kws={'color':'green'})
```

```python
df1.corr()
```

high correlation between budget v gross & votes v gross


```python
corr_mat = df1.corr()
sns.heatmap(corr_mat, annot = True)
plt.title('correlation matrix for numeric features')

plt.show()
```

```python
#numerize the categorical columns
df_num = df1
for col in df_num.columns:
    if (df_num[col].dtype == 'object'):
        df_num[col] = df_num[col].astype('category')
        df_num[col] = df_num[col].cat.codes
        
df_num
```

```python
#correlation matrix for all columns numerized
corr_mat = df_num.corr()
sns.heatmap(corr_mat, annot = True)
plt.title('correlation matrix for all features')

plt.show()
```

```python
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
corr_matrix = df_num.corr()
corr = corr_matrix.unstack()
corr
```

```python
sorted_pairs = corr.sort_values()
sorted_pairs
```

movie budgets & gross revenue have the highest correlation followed by votes and gross revenue
